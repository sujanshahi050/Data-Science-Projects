### DATA CLEANING

To clean all the tweets on 'text' column, i am using a python package called tweet-preprocessor. It removes URLs, Hashtags, Mentions, Reserved words, Emojis, Smileys from the text. Link for the package https://pypi.org/project/tweet-preprocessor/

```python
def clean_tweets(df,column_name):
    """
        Function to clean the text data 
        Input: df(DataFrame): dataframe object containing the column of tweets
               column_name (str) : Name of the column for tweets
        Output: df (DataFrame) dataframe object with cleaned tweets
    """
    def remove_punctuation(s):
        """
            Function to remove punctuation from a given string
            Input: s (str) String object 
            Output: s (str) String object with punctuation removed
        """
        s = s.translate(str.maketrans('', '', string.punctuation))
        return s
    
    for i in range(len(df)):
        s = df.loc[i,column_name]
        s = p.clean(s)
        s = remove_punctuation(s)
        df.loc[i,column_name] = s
    return df
    
```

```python
new_df = clean_tweets(df,'text')
```
