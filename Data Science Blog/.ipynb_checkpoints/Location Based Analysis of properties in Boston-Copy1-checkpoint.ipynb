```python
def cols_with_missing_values(df):
    """
        Function that finds columns with missing values and outputs a dictionary containing columns with missing 
        values as key and percentage of missing values as a value.
        
        INPUT: Takes in a dataframe object as an input
        OUTPUT: Returns a dictionary with column as a key and number of missing values from the column as a value
    
    """
    missing_values_col_list = {}
    total_rows = df.shape[0]
    for col in df.columns:
        count = sum(df[col].value_counts())
        if count < total_rows:
            missing_values_col_list[col] = (total_rows-count) 
    return missing_values_col_list

missing_values_cols = cols_with_missing_values(listings_df)
print((missing_values_cols))
```

DATA CLEANING


```python
def drop_cols_with_urls(df):
    """
        Function that drops all columns whose name has 'url' in it.
        
        INPUT: Takes in a dataframe object as an input
        OUTPUT: Returns nothing but drops all columns whose name contain 'url' from the dataframe
    
    """
    url_cols_list = []
    for col in df.columns:
        if 'url' in col:
            url_cols_list.append(col)
    return df.drop(url_cols_list,axis=1)

listings_df = drop_cols_with_urls(listings_df)          
```
